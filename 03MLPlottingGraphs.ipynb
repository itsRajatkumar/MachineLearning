<h3>Plotting X^3

```python
x=np.array([1,2,3,4,5])
y=x**3
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5,0.01)
y=x**3
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5,0.01)
y=x**3
plt.title("X^3 Graph")
plt.xlabel("X")
plt.ylabel("X^3")
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5)
y=x**3
plt.plot(x,y, color="red",marker="p",linewidth=.5)
plt.show()
```

Plotting X^4 , X^3 and X^2

```python
x=np.arange(0,5.01,0.01)
y1=x**2
y2=x**3
y3=x**4

plt.plot(x,y1,color="black", label ='X^2')
plt.plot(x,y2,color="blue", label='X^3')
plt.plot(x,y3,color="red", label='X^4')
plt.title("Power Graph")
plt.xlabel("X")
plt.ylabel("X Power")
plt.legend()

plt.show()
```

```python
x=np.arange(0,5.01,0.01)
y1=x**2
y2=x**3
y3=x**4

plt.plot(x,y1,color="black", label ='X^2')
plt.plot(x,y2,color="blue", label='X^3')
plt.plot(x,y3,color="red", label='X^4')
plt.axis([0,7,0,700])
plt.title("Power Graph")
plt.xlabel("X")
plt.ylabel("X Power")
plt.legend()

plt.show()
```

```python
x=np.arange(-5,5,0.01)
y=x**3
plt.title("X^3 Graph")
plt.xlabel("X")
plt.ylabel("X^3")
plt.plot(x,y)
plt.show()
```

<h1>Pie Charts

```python
size=[1,3,4,5,6]
plt.pie(size)
#plt.axis('equal')
plt.show()
```

```python
size=[1,3,4,5,6]
color=["black","purple","red","pink","blue"]
plt.pie(size,colors=color)
#plt.axis('equal')
plt.show()
```

```python
size=[1,3,4,5,6]
color=["black","purple","red","pink","blue"]
label=["A","B","C","D","E"]
plt.pie(size,colors=color,labels=label)
#plt.axis('equal')
plt.show()
```

```python
size=[1,3,4,5,6]
color=["green","purple","red","pink","blue"]
label=["A","B","C","D","E"]
plt.pie(size,colors=color,labels=label, autopct="%.2f%%")
#plt.axis('equal')
plt.show()
```

```python
size=[1,3,4,5,6]
expld=[.1,.2,0,0,.1]
color=["black","purple","red","pink","blue"]
label=["A","B","C","D","E"]
plt.pie(size,colors=color,labels=label,explode=expld)
#plt.axis('equal')
plt.show()
```

```python
size=[1,3,4,5,6]
expld=[.1,.2,0,0,.1]
color=["black","purple","red","pink","blue"]
label=["A","B","C","D","E"]
plt.pie(size,colors=color,labels=label,explode=expld,counterclock=False)
#plt.axis('equal')
plt.show()
```

```python
size=[1,3,4,5,6]
expld=[.1,.2,0,0,.1]
color=["black","purple","red","pink","blue"]
label=["A","B","C","D","E"]
plt.pie(size,colors=color,labels=label,explode=expld,counterclock=False, startangle=90)
#plt.axis('equal')
plt.show()
```
